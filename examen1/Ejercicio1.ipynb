**d)** Un cliente compra 50 paquetes de café ¿Cuál es la probabilidad que exactamente 20 de esas bolsas tengan menos de 195 gramos de café?

La probabilidad de ocurrencia de bolsas con menos de 195 gramos es de 0.08, que en 50 bolsas son 3.98. Ahora bien, 20/50 bolsas son 0.4 (40%), por lo que el 40% de la probabilidad de que las bolsas sean menor a 195 gramos parece ser la respuesta que buscamos, 0.032:

```r
p_menor195 = pnorm(195, mean = mu_nueva, sd = sigma)
p_menor195*0.4
```
