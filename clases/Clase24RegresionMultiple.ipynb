# Clase 24 - Regresión Múltiple

- Libro Applied Linear Regresion 3ed
- Usa el paquete del libro alr3


```r
library(alr3)
attach(cakes)
names(cakes)
```

```r
m = lm(Y ~ X1 + X2 + X1:X2 + I(X1^2) + I(X2^2), data=cakes)
summary(m)
# NUNCA limpiar variables a lo tonto viendo esta tabla. SALVO que únicamente queramos sacar UNO, sino no,
# porque sacar uno afecta al resto, asi que sería un error GARRAFAL sacar mas de uno.

# si queremos sacar uno vemos el pvalor
```

```r
residualPlot(m)
```

```r
plot(m)
```

```r
mmp(m)
```

```r
# si queremos ver si la varianza es constante
mmp(m, sd=TRUE)
```

```r
library(car)
ncvTest(m) # non constant variant test
```
