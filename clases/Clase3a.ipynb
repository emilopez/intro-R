## Grafico la potencia variando el delta

```r
a = 1
n = 10
alpha = 0.05
deltas = seq(0,3*a,length=100)
potencia = numeric(100)
umbral = qnorm(alpha, mean = 0, sd=a/sqrt(n))

for (j in 1:100){
    potencia[j] = pnorm(umbral, mean = 0 - deltas[j], sd=a/sqrt(n))
}
plot(deltas, potencia)
```

## Obtener el n
Para obtener un *n* determinado partiendo de una potencia requerida, fijando un delta específico

```r
delta = 1
potencia_requerida = 0.8
for (n in 3:100){
    umbral = qnorm(alpha, mean = 0, sd=a/sqrt(n))
    potencia = pnorm(umbral, mean = 0 - delta, sd=a/sqrt(n))
    if (potencia > potencia_requerida){
        print(n)
        break
    }
}
n
```
