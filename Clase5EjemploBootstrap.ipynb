```r
library(MASS)
n = 20
muestra = rnorm(n, mean=0, sd = 1)
t = mean(muestra, trim=0.1)
```

```r
B = 1000
tboot = numeric(B)
for (b in 1:B){
    muestra_boot = sample(muestra, n, replace = TRUE) # con esto le decimos que lo haga con sustitucion
    tboot[b] = mean(muestra_boot, trim=0.1)
}
```

```r
truehist(tboot)
```

```r
# como veo que no es simetrica, esta sesgada, entonces calculo los cuantiles, si era simetrica podÃ­a calcular 
# el IC con la formula de t +/- z_aplha/2 * sd(tboot) =>
# IC:
#   simetrica     -> IC: (t +/- z_aplha/2 * sd(tboot) )
#   no simetrica  -> IC: quantile(tboot, c(alpha/2, 1-alpha/2))
alpha = 0.05
quantile(tboot, c(alpha/2, 1-alpha/2))
```
